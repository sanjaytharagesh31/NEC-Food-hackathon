```python
#separate the dataset
df_ideal = df.loc[:99]
df_moderate = df.loc[100:499]
df_worst = df.loc[500:]
print(df_ideal)
print(df_moderate)
print(df_worst)
```

```python
def shuf_col_ideal(colnam):
    a = df_ideal[colnam]
    a = a.sample(frac=1)
    a.reset_index(inplace=True, drop=True)
    return a
```

```python
def shuf_col_moderate(colnam):
    a = df_moderate[colnam]
    a = a.sample(frac=1)
    a.reset_index(inplace=True, drop=True)
    return a
```

```python
def shuf_col_worst(colnam):
    a = df_worst[colnam]
    a = a.sample(frac=1)
    a.reset_index(inplace=True, drop=True)
    return a
```

```python
#shuffle ideal dataset
a = shuf_col_ideal('temperature')
b = shuf_col_ideal('humidity')
c = shuf_col_ideal('top_ice_accepted')
d = shuf_col_ideal('water_sprinkle_accepted')
e = shuf_col_ideal('ethylene_ppm')
f = shuf_col_ideal('C02')
g = shuf_col_ideal('Oxygen')
h = shuf_col_ideal('shelf_life')

dict_ideal = {'temperature': a, 'humidity': b, 'top_ice_accepted': c, 'water_sprinkle_accepted': d, 'ethylene_ppm': e,'c02': f,'oxygen': g, 'shelf_life': h}  
df_ideal_shuf = pd.DataFrame(dict_ideal)
df_ideal_shuf
```

```python
#shuffle moderate dataset
a = shuf_col_moderate('temperature')
b = shuf_col_moderate('humidity')
c = shuf_col_moderate('top_ice_accepted')
d = shuf_col_moderate('water_sprinkle_accepted')
e = shuf_col_moderate('ethylene_ppm')
f = shuf_col_moderate('C02')
g = shuf_col_moderate('Oxygen')
h = shuf_col_moderate('shelf_life')

dict_moderate = {'temperature': a, 'humidity': b, 'top_ice_accepted': c, 'water_sprinkle_accepted': d, 'ethylene_ppm': e,'c02': f,'oxygen': g, 'shelf_life': h}  
df_moderate_shuf = pd.DataFrame(dict_moderate)
df_moderate_shuf
```

```python
#shuffle worst dataset
a = shuf_col_worst('temperature')
b = shuf_col_worst('humidity')
c = shuf_col_worst('top_ice_accepted')
d = shuf_col_worst('water_sprinkle_accepted')
e = shuf_col_worst('ethylene_ppm')
f = shuf_col_worst('C02')
g = shuf_col_worst('Oxygen')
h = shuf_col_worst('shelf_life')

dict_worst = {'temperature': a, 'humidity': b, 'top_ice_accepted': c, 'water_sprinkle_accepted': d, 'ethylene_ppm': e,'c02': f,'oxygen': g, 'shelf_life': h}  
df_worst_shuf = pd.DataFrame(dict_worst)
df_worst_shuf
```

```python
#concatenate ideal, moderate, worst dfs
frames = [df_ideal_shuf, df_moderate_shuf, df_worst_shuf]
result = pd.concat(frames)
result
```

```python
#shuffle complete dataset

data = result.sample(frac=1)
data.reset_index(inplace=True, drop=True)
data.to_csv(DATA_PATH+'yeetFinal.csv', index=False)
data
```
